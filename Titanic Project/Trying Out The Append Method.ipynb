```python
transtrain = traindata.drop("Survived", axis=1) #to drop label to be predicted so shape is same as test
Y_train = traindata["Survived"] #to keep dropped label to be added to transformed trained data later
#Drop header in testdata
#testheader = testdata.loc[0]
#finaltest = testdata[1:]
```

```python
finaldata = transtrain.append(testdata) #combining test data and train data
```

```python
finaldata.shape
```

```python
#performing transformations to the combined data set
meanage = round(finaldata['Age'].mean(),1)
finaldata.Age.fillna(meanage, inplace = True)
finaldata = pd.get_dummies(finaldata, columns=["Pclass","Embarked","Sex"])
finaldata.drop(['Cabin', 'Name', 'Ticket','Fare' ], axis=1, inplace=True)
```

```python
finaldata
```

```python
#to check effect of transformations
finaldata.isnull().sum()*100/len(finaldata)
```

```python
#find out where test data begins from
finaldata.iloc[891]
```

```python
#split data back into test and train
X_train = finaldata.iloc[:891,:]
finaltest = finaldata.iloc[891:,:]

#Remember to add initially dropped survive column back to train data
```

```python
#adding survived back to train data
finaltrain = X_train.join(Y_train)
```
